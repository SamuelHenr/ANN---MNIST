```python
# First, we are gonna get the features vector from the file
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torchvision
from torchvision.ops import MLP
import matplotlib.pyplot as plt
import math

df = np.array(pd.read_csv('data_tp1', sep=','))
train_size = math.floor(len(df)*0.8)

# First column is the label
y = df[:,0]
y_train = y[0:train_size]
y_test = y[(train_size+1):len(df)]

# Separating the data
x = np.delete(df, 0, 1)
x_train = x[0:train_size]
x_test = x[(train_size+1):len(df)]

# Getting Tensor data structure
train_data = torch.tensor(x_train, dtype=torch.float32)
test_data = torch.tensor(x_test, dtype=torch.float32)
label = torch.tensor(y_train)
test_label = torch.tensor(y_test)
```

```python
# Implementing Gradient Descent
model = MLP(in_channels=784, hidden_channels=[25, 10], activation_layer=torch.nn.Sigmoid)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.5)

num_epochs = 100
loss_GD_25 = [] # stores the loss for each epoch
for n in range(num_epochs):
    # Forward pass
    y_pred = model(train_data)
    # Calculating the loss
    loss = loss_fn(y_pred, label)
    loss_GD_25.append(loss.item())
    # Backward pass (gradient calculation is done here)
    optimizer.zero_grad()
    loss.backward()
    # Updating the weigths
    optimizer.step()
```

```python
# Implementing Stochastic Gradient Descent
model = MLP(in_channels=784, hidden_channels=[25, 10], activation_layer=torch.nn.Sigmoid)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.5)

num_epochs = 10
loss_SGD_25 = [] # stores the loss for each epoch
for n in range(num_epochs):
    # Forward pass
    y_pred = model(train_data)
    # Calculating the loss
    loss = loss_fn(y_pred, label)
    loss_SGD_25.append(loss.item())
    for X, Y in zip(train_data, label):
        y_hat = model(X)
        loss = loss_fn(y_hat, Y)
        # Backward pass (gradient calculation is done here)
        optimizer.zero_grad()
        loss.backward()
        # Updating the weigths
        optimizer.step()
```

```python
# Implementing Mini-Batch Gradient Descent
model = MLP(in_channels=784, hidden_channels=[25, 10], activation_layer=torch.nn.Sigmoid)
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=0.5)

batch = 10
num_epochs = 20
loss_BGD_25 = [] # stores the loss for each epoch
for n in range(num_epochs):
    # Forward pass
    y_pred = model(train_data)
    # Calculating the loss
    loss = loss_fn(y_pred, label)
    loss_BGD_25.append(loss.item())

    start = 0
    finish = batch
    while finish <= len(label):
        # Training range of values
        X = train_data[start:finish]
        Y = label[start:finish]
        y_hat = model(X)
        loss = loss_fn(y_hat, Y)
        # Backward pass (gradient calculation is done here)
        optimizer.zero_grad()
        loss.backward()
        # Updating the weigths
        optimizer.step()
        start=start+batch
        finish=finish+batch
```
